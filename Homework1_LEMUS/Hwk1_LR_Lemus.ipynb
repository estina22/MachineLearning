```python
import os
import cv2
import random
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,accuracy_score
```

```python
dir_train ='/Users/eslemus/Desktop/ML/Anticl/images/Train'
dir_test ='/Users/eslemus/Desktop/ML/Anticl/images/Test'
label_type = ['Class1','Class2']

img_size = 150
im_l =240
im_b =320
training_data = []
```

```python
def create_data(dir_train,label_type):
    for seismic_type in label_type:
        path = os.path.join(dir_train,seismic_type)
        classnum = label_type.index(seismic_type)
        for img in os.listdir(path):
            img_arr = cv2.imread(os.path.join(path,img),0)
            print(img)
            if img_arr is not None:
            	new_arr = cv2.resize(img_arr,(img_size,img_size))
            	training_data.append([new_arr,classnum])
            
create_data(dir_train,label_type)
print(len(training_data))
random.shuffle(training_data)
```

```python
x=[]
y=[]
for feature , label in training_data:
    x.append(feature)
    y.append(label)
print(len(x),len(y))
X = np.array(x)

print(X.shape)
```

```python
train_imgs_scaled = X.astype('float32')
train_imgs_scaled /= 255

x_train_new = train_imgs_scaled.reshape(len(train_imgs_scaled),train_imgs_scaled.shape[1]*train_imgs_scaled.shape[2])
```

```python
logisticRegr = LogisticRegression()
logisticRegr.fit(x_train_new, y)

create_data(dir_test,label_type)
print(len(training_data))
random.shuffle(training_data)

test_x=[]
test_y=[]

for feature , label in training_data:
    test_x.append(feature)
    test_y.append(label)
print(len(test_x),len(test_y))
tx = np.array(test_x)

print(X.shape)

test_imgs_scaled = tx.astype('float32')
test_imgs_scaled /= 255

x_test_new = test_imgs_scaled.reshape(len(test_imgs_scaled),test_imgs_scaled.shape[1]*test_imgs_scaled.shape[2])
```

```python
from sklearn.metrics import confusion_matrix
y_pred = logisticRegr.predict(x_test_new)
cm = confusion_matrix(test_y, y_pred)

print("Accuracy Score -> ",accuracy_score(test_y, y_pred)*100)
print(classification_report(test_y,y_pred))
print(cm)

#print(y_pred)
```
